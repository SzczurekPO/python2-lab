```python
# 1. Oblicz medianę wynagrodzenia
mediana = df['wynagrodzenie'].median()
print(f"Mediana wynagrodzenia: {mediana}")

# 2. Wypełnij NaN medianą
df['wynagrodzenie'] = df['wynagrodzenie'].fillna(mediana)

# Sprawdzenie efektu
print("\n=== WYNAGRODZENIE PO UZUPEŁNIENIU ===")
print(df['wynagrodzenie'].head(10))

# 3. Oblicz średnią ocena_roczna
srednia = round(df['ocena_roczna'].mean(), 2)
print(f"\nŚrednia ocena roczna: {srednia}")

# 4. Wypełnij NaN średnią
df['ocena_roczna'] = df['ocena_roczna'].fillna(srednia)

# Sprawdzenie efektu
print("\n=== OCENA ROCZNA PO UZUPEŁNIENIU ===")
print(df['ocena_roczna'].head(10))

# 5. Sprawdź ile NaN zostało
print("\n=== POZOSTAŁE BRAKI ===")
print(df.isna().sum())
```

```python
# Zacznij od świeżej kopii
df2 = pd.DataFrame(data)  # ta sama definicja 'data' co wyżej

# Najpierw napraw wynagrodzenie (tak samo jak w Ćw. 1b)
df2['wynagrodzenie'] = df2['wynagrodzenie'].replace('brak', np.nan)
df2['wynagrodzenie'] = pd.to_numeric(df2['wynagrodzenie'], errors='coerce')
```

```python
# 1. Liczba pełnych duplikatów
print("=== PEŁNE DUPLIKATY ===")
print(df2.duplicated().sum())

# 2. Wyświetlenie pełnych duplikatów
print("\n=== ZDUPLIKOWANE WIERSZE ===")
print(df2[df2.duplicated()])

# 3. Duplikaty wg id_pracownika
print("\n=== DUPLIKATY WG ID ===")
print(df2.duplicated(subset=['id_pracownika']).sum())

# 4. Liczba wystąpień każdego ID
print("\n=== LICZBA WYSTĄPIEŃ ID ===")
print(df2['id_pracownika'].value_counts().sort_index())
```

```python
df2 = df2.drop_duplicates().reset_index(drop=True)

print("\n=== NOWY KSZTAŁT DF ===")
print(df2.shape)
```

```python
# 1. Typ przed konwersją
print("\n=== TYP PRZED ===")
print(df2['data_zatrudnienia'].dtype)

# 2. Konwersja na datetime
df2['data_zatrudnienia'] = pd.to_datetime(
    df2['data_zatrudnienia'],
    errors='coerce'
)
```

```python
# Czyszczenie kolumny dzial
df2['dzial'] = df2['dzial'].str.strip().str.title()

# 1. Zatrudnieni od 2021 roku
df2['rok_zatrudnienia'] = df2['data_zatrudnienia'].dt.year
po_2021 = df2[df2['rok_zatrudnienia'] >= 2021]

print("\n=== ZATRUDNIENI OD 2021 ===")
print(len(po_2021))

# 2. Wynagrodzenie > 6000
powyzej_6000 = df2[df2['wynagrodzenie'] > 6000]

print("\n=== WYNAGRODZENIE > 6000 ===")
print(len(powyzej_6000))

# 3. Liczba pracowników w dziale IT
it_count = df2[df2['dzial'] == 'It']

print("\n=== LICZBA PRACOWNIKÓW W IT ===")
print(len(it_count))
```
