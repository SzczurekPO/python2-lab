```python
# Dane
dane = np.array([
    [120, 145, 167],   # Produkt A
    [89,  102, 95],    # Produkt B
    [234, 267, 289],   # Produkt C
    [56,  78,  91],    # Produkt D
])

produkty = ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D']
kwartaly = ['Q1', 'Q2', 'Q3']

# 1. Sprzedaż Produktu A w Q2
print("Produkt A, Q2:", dane[0, 1])

# 2. Cała sprzedaż Produktu C
print("Produkt C (wszystkie kwartały):", dane[2])

# 3. Sprzedaż wszystkich produktów w Q1
print("Wszystkie produkty, Q1:", dane[:, 0])

# 4. Produkty B i C w Q2 i Q3
print("Produkty B i C, Q2-Q3:")
print(dane[1:3, 1:3])
```

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
srednia = ceny.mean()
najdrozszy = ceny.max()
najtanszy = ceny.min()

idx_max = ceny.argmax()
idx_min = ceny.argmin()

print("Średnia cena:", srednia)
print("Najdroższy:", nazwy[idx_max], "-", najdrozszy)
print("Najtańszy:", nazwy[idx_min], "-", najtanszy)
```

```python
wartosci = ceny * ilosci_magazyn
laczna_wartosc = wartosci.sum()

print("Wartość każdego produktu:", wartosci)
print("Łączna wartość magazynu:", laczna_wartosc)
```

```python
ceny_rabat = ceny * 0.85

ceny_vat = ceny * 1.23

ile_drogich = (ceny > 1000).sum()

print("Ceny po rabacie:", ceny_rabat)
print("Ceny z VAT:", ceny_vat)
print("Liczba produktów > 1000 zł:", ile_drogich)
```

```python
nazwy_np = np.array(nazwy)

maska_oceny = oceny_klientow >= 4.5
print("Produkty z oceną >= 4.5:")
print(nazwy_np[maska_oceny])
print(ceny[maska_oceny])

maska_tanie = ceny < 200
print("\nProdukty < 200 zł:")
print(nazwy_np[maska_tanie])
print(ilosci_magazyn[maska_tanie])
```

```python
nazwy_np = np.array(nazwy)
wartosci = ceny * ilosci_magazyn

maska_niski = ilosci_magazyn < 30

produkty_niski = nazwy_np[maska_niski]
wartosc_niski = wartosci[maska_niski].sum()

procent = (wartosc_niski / wartosci.sum()) * 100

print("Produkty z niskim stanem:", produkty_niski)
print("Łączna wartość:", wartosc_niski)
print("Udział %:", procent)
```

## Wnioski z benchmarku
- NumPy jest **24×** szybszy w mnożeniu
- NumPy jest **31×** szybszy w sumowaniu
- NumPy jest **24×** szybszy w filtrowaniu
- Wniosek: przy dużych danych **zawsze** używaj NumPy zamiast list

```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
