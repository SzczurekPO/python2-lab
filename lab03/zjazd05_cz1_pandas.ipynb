```python
pensje = pd.Series({
    'Anna': 5500,
    'Jan': 7200,
    'Ewa': 4800,
    'Marek': 9100,
    'Kasia': 6300
})

print("Pensje:\n", pensje)
```

```python
najwiecej = pensje.idxmax()
print("\nNajwięcej zarabia:", najwiecej)

srednia = pensje.mean()
print("Średnia pensja:", srednia)

powyzej_sredniej = pensje[pensje > srednia]
print("Powyżej średniej:\n", powyzej_sredniej)

ponizej_6000 = (pensje < 6000).sum()
print("Liczba osób poniżej 6000:", ponizej_6000)
```

```python
podwyzka = pensje * 1.10

premia = pensje + 500

nowe_pensje = (pensje * 1.10) + 500

print("\nNowe pensje:\n", nowe_pensje)

ranking = nowe_pensje.sort_values(ascending=False)
print("\nRanking nowych pensji:\n", ranking)
```

```python
import pandas as pd

sklep = pd.DataFrame({
    'produkt': ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Monitor'],
    'cena': [3500, 1800, 2500, 350, 2200],
    'stan_magazynowy': [45, 120, 200, 500, 75],
    'kategoria': ['Komputery', 'Mobilne', 'Mobilne', 'Akcesoria', 'Komputery']
})
```

```python
print(sklep.shape)

print(sklep.columns)

print(sklep.dtypes)

print(sklep.head(3))
```

```python
sklep['wartosc_magazynu'] = sklep['cena'] * sklep['stan_magazynowy']

sklep['cena_z_vat'] = sklep['cena'] * 1.23

print(sklep)
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)
```

```python
print(penguins.head(10))
print(penguins.shape)
print(penguins.info())
```

```python
print(penguins.describe())
print(penguins.describe(include='all'))
print(penguins['body_mass_g'].median())
```

```python
print(penguins['species'].value_counts())
print(penguins['island'].value_counts())
print(penguins['sex'].value_counts())
print(penguins['species'].value_counts(normalize=True))
```

```python
print(penguins.isna().sum())
print(penguins['sex'].isna().sum() / len(penguins) * 100)
```

```python
import pandas as pd

tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
print(tips['total_bill'].mean())

print(tips['tip_pct'].mean())

print(tips['day'].value_counts())

print(tips['smoker'].value_counts())

print(tips['tip'].describe())

print(tips['time'].value_counts(normalize=True))

print(tips['size'].nunique())
```
